```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Equations on the same plot

```python
k=4

n = np.arange(0.1, 7.1, 0.1)
n_2 = n ** 2
n_3 = n ** 3
n_poly = n ** k
n_log = np.log2(n)

plt.figure(figsize=(15, 15))
plt.plot(n, n_log, 'b', label="Logarithmic")
plt.plot(n, n, 'r', label="Linear")
plt.plot(n, n_2, 'g', label="Quadratic")
plt.plot(n, n_3, 'y', label="Cubic")
plt.plot(n, n_poly, 'm', label="Polynomial")
plt.xlabel('n')
plt.ylabel('Time in seconds')
plt.legend(loc='upper right')
plt.ylim((0,50))
plt.show()
```

#### Equations on separate plots with n=7 and k=4

```python
figure, axis = plt.subplots(2, 3, figsize=(15,15))

axis[0, 0].plot(n, n_log, 'b')
axis[0, 0].set_title("Logarithmic")
axis[0, 0].text(7, np.log2(7), '({}, {})'.format(7, round(np.log2(7),2)))
axis[0, 0].set_xlabel('n')
axis[0, 0].set_ylabel('(log n)')

axis[0, 1].plot(n, n, 'r')
axis[0, 1].set_title("Linear")
axis[0, 1].text(7, 7, '({}, {})'.format(7, 7))
axis[0, 1].set_xlabel('n')
axis[0, 1].set_ylabel('(n)')
  
axis[0, 2].plot(n, n_2, 'g')
axis[0, 2].set_title("Quadratic")
axis[0, 2].text(7, 7 ** 2, '({}, {})'.format(7, 7 ** 2))
axis[0, 2].set_xlabel('n')
axis[0, 2].set_ylabel('(n^2)')

axis[1, 0].plot(n, n_3, 'y')
axis[1, 0].set_title("Cubic")
axis[1, 0].text(7, 7 ** 3, '({}, {})'.format(7, 7 ** 3))
axis[1, 0].set_xlabel('n')
axis[1, 0].set_ylabel('(n^3)')

axis[1, 1].plot(n, n_poly, 'm')
axis[1, 1].set_title("Polynomial")
axis[1, 1].text(7, 7 ** 4, '({}, {})'.format(7, 7 ** 4))
axis[1, 1].set_xlabel('n')
axis[1, 1].set_ylabel('(n^k)')

figure.delaxes(axis[1,2])

plt.tight_layout()

plt.show()
```
